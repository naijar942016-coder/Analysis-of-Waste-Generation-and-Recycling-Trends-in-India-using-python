### **Skewness Check & Transformation**

```python
import numpy as np

num_cols = df.select_dtypes(include='number').columns.tolist()
num_cols = [c for c in num_cols if c != 'Year']

for col in num_cols:
    skew = df[col].skew()
    if abs(skew) >= 1.0:
        df[f'{col}_log'] = np.log1p(df[col])
        print(f"Transformed: {col}")
```

trend analysis and comparison studies.

c.**Correlation Heatmap**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
num_cols = df.select_dtypes(include='number').columns.tolist()
print("Numeric columns found:", num_cols)
corr = df[num_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
short_labels=[col[:10] for col in num_cols]

plt.figure(figsize=(12, 9))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            xticklabels=short_labels, yticklabels=short_labels,
            linewidths=0.5)
plt.title('Correlation Heatmap – All Numerical Features',
          fontsize=14, fontweight='bold')
plt.xticks(rotation=40, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.show()
```

Key correlation findings:
- **Waste_Generated ↔ Waste_Recycled: ~1.0** — as expected, cities
  that generate more waste also recycle more in absolute volume
- **Waste_Generated ↔ Total_Daily_Cost: ~1.0** — higher waste volume
  directly drives higher total management cost (obvious but confirmed)
- **Waste_Generated ↔ Recycling_Rate: -0.03** — near zero — volume
  of waste does NOT affect recycling rate percentage
- **Efficiency_Score ↔ Recycling_Rate: ~0.0** — governance score
  alone does NOT predict recycling performance

Most important business takeaway: **No single factor strongly predicts

### **Multivariate Analysis**

a.**Recycling Rate by Waste Type & Disposal Method (Grouped Barplot)**

```python
pivot13 = df.groupby(['Waste_Type','Disposal_Method'])['Recycling_Rate'].mean().reset_index()

plt.figure(figsize=(13, 6))
ax13 = sns.barplot(data=pivot13, x='Waste_Type', y='Recycling_Rate',
                   hue='Disposal_Method',)
plt.title('Avg Recycling Rate by Waste Type & Disposal Method',
          fontsize=14, fontweight='bold')
plt.xlabel('Waste Type', fontsize=11)
plt.ylabel('Avg Recycling Rate (%)', fontsize=11)
plt.legend(title='Disposal Method', fontsize=9, title_fontsize=10)
plt.axhline(df['Recycling_Rate'].mean(), color='black',
            linestyle='--', linewidth=1.2, label='Overall Mean')
plt.show()
```

Across all waste types, disposal methods perform **very similarly**
— no single combination of waste type + disposal method dramatically
outperforms others. This reinforces that:
- The **system-level approach** (not just waste type or method choice)
  determines recycling success
- For **Plastic waste**, Landfill-paired systems perform marginally
  better — suggesting areas using landfills have stronger parallel
  plastic collection and recycling systems
- For **E-Waste and Hazardous**, no disposal method stands out —
  these need **specialised treatment facilities** regardless of
  disposal route, which India currently lacks at scale.

b. **City × Year Recycling Rate Heatmap (Pivot Table)**

```python
pivot14 = df.pivot_table(index='City', columns='Year',
                          values='Recycling_Rate',)

plt.figure(figsize=(12, 14))
sns.heatmap(pivot14, annot=True, fmt='.1f', cmap='RdYlGn',
            linewidths=0.4, annot_kws={'size': 8},
            cbar_kws={'label': 'Avg Recycling Rate (%)'})
plt.title('City × Year Recycling Rate Heatmap\n(Green = Higher, Red = Lower)',
          fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=11)
plt.ylabel('City', fontsize=11)
plt.xticks(fontsize=10)
plt.yticks(fontsize=9, rotation=0)
plt.show()
```

The heatmap reveals **city-specific trajectories** over 5 years:
- **Dark green cells** = consistently high recyclers across years
  (Indore, Varanasi, Ahmedabad) — these cities have built
  sustainable, year-stable recycling systems
- **Dark red cells** = consistently low recyclers (Gwalior, Kanpur)
  — these cities have structural problems not solved by time alone
- **Colour shift from red to green** across years = improving city
  — these are the cities where policy interventions are working

Business action: Cities with **stable red across all 5 years**
(not just 1 bad year) need emergency infrastructure upgrades —
their problem is systemic, not situational.

# **Final Overall Insights:**



Across 34 Indian cities and five years of data, the overall picture is one of a system that performs
moderately well on average but is highly inconsistent underneath.

On average, just over half of all waste
generated each day (57.08%) gets recycled — meaning the other 43% ends up unprocessed, in landfills,
or handled in less sustainable ways. But this national average hides a wide spread: some cities are
recycling nearly two-thirds of their waste, while others recycle barely half.

What is more surprising is what does not explain this gap. It is tempting to assume that cities spending
more money, running more awareness campaigns, or scoring higher on official efficiency ratings would
naturally recycle more.

The data does not support that assumption — none of these factors show a
meaningful relationship with recycling performance. This points to a deeper, more human explanation:
recycling success in India seems to depend on how well a city executes the basics — consistent
collection, simple segregation habits, and reliable infrastructure — rather than how much it spends or how
many campaigns it runs on paper.

In short, the data tells a story of untapped, fixable potential. The tools and spending are often already in
place; what is missing in low-performing cities is consistent on-the-ground execution, which can be
learned from cities that are already getting it right.
